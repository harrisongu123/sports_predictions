```python
import pandas as pd
from sklearn.model_selection import train_test_split
from keras import models
from keras.models import Sequential
from keras import layers
from keras.layers import Dense, Dropout, Activation
from keras import optimizers
from keras import regularizers
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error
import talos
```

```python
df = pd.read_csv('/content/drive/MyDrive/avg_by_game')
```

```python
df.head()
```

```python
data = df.set_index(df['Unnamed: 0'], drop=True)
data.index.rename('GameID', inplace=True)
data.drop(columns='Unnamed: 0', inplace=True)
```

```python
y = data['Home Win By']
x = data.drop(columns='Home Win By')

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=42)
```

```python
def dense_network(x_train, y_train, x_test, y_test, params):

    # we build the model exactly like we would normally do it
    model = Sequential()

    # hidden layers
    model.add(Dense(params['nodes1'], input_dim=29, activation=params['activation 1']))
    model.add(Dense(round(params['nodes1']/2), activation=params['activation 2']))
    model.add(Dense(round(params['nodes1']/4), activation=params['activation 3']))
    
    # output layer
    model.add(Dense(1, activation='linear'))

    model.compile(loss='mse', optimizer=params['optimizer'], metrics=["mean_squared_error"])
    
    out = model.fit(x_train, y_train, 
                        validation_data=(x_test, y_test),
                        batch_size=300,
                        epochs=200,
                        verbose=0)

    return out, model
```

```python
params = { 'nodes1': [100, 500, 1000],
          'activation 1': ['linear', 'relu'],
          'activation 2': ['linear', 'relu'],
          'activation 3': ['linear', 'relu'],
          'optimizer': ['adam', 'sgd']
          }
```

```python
results = talos.Scan(x_train.values, y_train.values, params=params, model=dense_network, experiment_name='grid')
results.best_model("mean_squared_error")
```
