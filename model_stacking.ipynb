```python
def similar_picks2(prediction1, prediction2, vegas, result):
    games_bet = 0
    bet_win = 0
    for i in range(len(result)):
        if (((result[i]>vegas[i]) & (prediction1[i]>vegas[i])) & ((result[i]>vegas[i]) & (prediction2[i]>vegas[i]))):
            games_bet += 1
            bet_win += 1
        elif (((result[i]<vegas[i]) & (prediction1[i]<vegas[i])) & ((result[i]<vegas[i]) & (prediction2[i]<vegas[i]))):
            games_bet += 1
            bet_win += 1
        elif (((result[i]>vegas[i]) & (prediction1[i]<vegas[i])) & ((result[i]>vegas[i]) & (prediction2[i]<vegas[i]))):
            games_bet += 1
        elif (((result[i]<vegas[i]) & (prediction1[i]>vegas[i])) & ((result[i]<vegas[i]) & (prediction2[i]>vegas[i]))):
            games_bet += 1
    bet_rate = 100*(games_bet/len(result))
    win_rate = 100*(bet_win/games_bet)
    print('You would bet {}% of games, and win {}% of the time'.format(bet_rate, win_rate))
```
