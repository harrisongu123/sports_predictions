```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
%matplotlib inline
plt.style.use('seaborn-darkgrid')
```

```python
df = pd.read_csv('test_results')
```

```python
df.set_index('GameID', drop=True, inplace=True)
```

```python
df
```

```python
lr_bet = []
gb_bet = []
rf_bet = []
nn_bet  = []
actual = []
columns = ['Linreg Prediction', 'GB Prediction', 'RF Prediction', 'NN Predictions', 'Actual']
lists = [lr_bet, gb_bet, rf_bet, nn_bet, actual]
for x, column in enumerate(columns):
    for i in range(len(df)):
        if df[column][i] > df['Vegas Prediction'][i]:
            lists[x].append('H')
        elif df[column][i] < df['Vegas Prediction'][i]:
            lists[x].append('A')
        elif df[column][i] == df['Vegas Prediction'][i]:
            lists[x].append('T')
    df['{} H/A'.format(columns[x])] = lists[x]
```

```python
df
```

```python
lr_result = []
gb_result = []
rf_result = []
nn_result  = []

columns1 = ['Linreg Prediction H/A', 'GB Prediction H/A', 'RF Prediction H/A', 'NN Predictions H/A']
lists1 = [lr_result, gb_result, rf_result, nn_result]
for x, column in enumerate(columns1):
    for i in range(len(df)):
        if df[column][i] == df['Actual H/A'][i]:
            lists1[x].append('W')
        elif df[column][i] != df['Actual H/A'][i]:
            lists1[x].append('L')

    df['{} Result'.format(columns[x])] = lists1[x]
```

```python
df.drop(df[df['Actual H/A']=='T'].index, inplace=True)
```

```python
def analyze_distance(predictions, results, vegas):
    win_diff = []
    lose_diff = []
    for i in range(len(predictions)):
        if results[i] == 'W':
            win_diff.append(abs(predictions[i]-vegas[i]))
        elif results[i] == 'L':
            lose_diff.append(abs(predictions[i]-vegas[i]))
    
    accuracy = []
    for i in range(1,21):
        win_count = 0
        lose_count = 0
        for win in win_diff:
            if ((win <= i) & (win > i-1)):
                win_count += 1
        for loss in lose_diff:
                if ((loss <= i) & (loss > i-1)):
                    lose_count += 1
        accuracy.append((100*win_count)/(win_count+lose_count))
    
    fig, [[ax1, ax2], [ax3, ax4]] = plt.subplots(nrows=2, ncols=2, figsize=(20, 10))
    
    ax1.scatter(range(len(win_diff)), win_diff, label='Wins')
    ax1.scatter(range(len(lose_diff)), lose_diff, label='Losses')
    ax1.set_title('Distance from Vegas')
    ax1.set_ylabel('Distance')
    ax1.set_xlim((0,len(lose_diff))) 
    ax1.legend()
    
    ax2.boxplot([win_diff, lose_diff],
                     vert=False,  # vertical box alignment
                     patch_artist=True,
                     notch=True,# fill with color
                     labels=['Win', 'Lose'])  # will be used to label x-ticks
    # prettify and label
    ax2.set_title('Distance From Vegas by Win/Lose')
    ax2.xaxis.grid(True)
    ax2.set_xlabel('Distance')
#     ax2.set_xlim(0,20)
    
    ax3.hist(win_diff, bins=20, density=False, alpha=.7, label='Win', range=(0,20))
    ax3.hist(lose_diff, bins=20, density=False, alpha=.7, label='Lose', range=(0,20))
    ax3.set_title('Distance Overlay')
    ax3.set_xlabel('Distance')
    ax3.set_ylabel('# of Times')
    ax3.legend()

    ax4.plot(accuracy)
    ax4.set_title('Accuracy by Distance from Vegas')
    ax4.xaxis.grid(True)
    ax4.yaxis.grid(True)
    ax4.set_xlabel('Distance')
    ax4.set_ylabel('Win %')
```

```python
analyze_distance(df['Linreg Prediction'], df['Linreg Prediction Result'], df['Vegas Prediction'])
```

```python
def accuracy(win_diff, lose_diff):
    accuracy = []
    for i in range(1,21):
        win_count = 0
        lose_count = 0
        for win in win_diff:
            if ((win <= i) & (win > i-1)):
                win_count += 1
        for loss in lose_diff:
                if ((loss <= i) & (loss > i-1)):
                    lose_count += 1
        accuracy.append((100*win_count)/(win_count+lose_count))
    
    return accuracy
```

```python
def get_diff(predictions, vegas):
    diff = []
    for i in range(len(predictions)):
        diff.append(abs(predictions[i]-vegas[i]))
    
    return diff
```

```python
df['LR Diff'] = get_diff(df['Linreg Prediction'], df['Vegas Prediction'])
df['GB Diff'] = get_diff(df['GB Prediction'], df['Vegas Prediction'])
df['RF Diff'] = get_diff(df['RF Prediction'], df['Vegas Prediction'])
df['NN Diff'] = get_diff(df['NN Predictions'], df['Vegas Prediction'])
```

```python
lr_acc = accuracy(df[df['Linreg Prediction Result']=='W']['LR Diff'], df[df['Linreg Prediction Result']=='L']['LR Diff'])
gb_acc = accuracy(df[df['GB Prediction Result']=='W']['GB Diff'], df[df['GB Prediction Result']=='L']['GB Diff'])
rf_acc = accuracy(df[df['RF Prediction Result']=='W']['RF Diff'], df[df['RF Prediction Result']=='L']['RF Diff'])
nn_acc = accuracy(df[df['NN Predictions Result']=='W']['NN Diff'], df[df['NN Predictions Result']=='L']['NN Diff'])
all_acc = accuracy(win_dist, lose_dist)
```
